## კონცეპტუალური ამოცანები

#### 1) შემდეგი დავალებები განეეკუთვნებიან რეგრესიას თუ კლასიფიკაციას:
- ჩვენ ვაგროვებთ მონაცემთა ერთობლიობას აშშ-ში 500 საუკეთესო ფირმის შესახებ. თითოეული ფირმისთვის ჩვენ აღვნიშნავთ მოგებას, დასაქმებულთა რაოდენობას, ინდუსტრიას და აღმასრულებელი დირექტორის ხელფასს. ჩვენ დაინტერესებული ვართ გავიგოთ, რა ფაქტორები მოქმედებს აღმასრულებელი დირექტორის ხელფასზე
- ჩვენ განვიხილავთ ახალი პროდუქტის გამოშვებას და გვინდა ვიცოდეთ, იქნება ის წარმატებული თუ წარუმატებული. ჩვენ ვაგროვებთ მონაცემებს 20 მსგავსი პროდუქტის შესახებ, რომლებიც ადრე იყო გამოშვებული. თითოეული პროდუქტისთვის ჩვენ ვაკეთებთ შემდეგ ჩანაწერებს: წარმატებული იყო თუ წარუმატებელი, პროდუქტისთვის დაწესებული ფასი, მარკეტინგული ბიუჯეტი, კონკურენციის ფასი და ათი სხვა ცვლადი.
- ჩვენ დაინტერესებული ვართ ვიწინასწარმეტყველოთ აშშ დოლარი/ევროს კურსის პროცენტული ცვლილება მსოფლიო საფონდო ბირჟებზე ყოველკვირეულ ცვლილებებთან მიმართებაში. აქედან გამომდინარე, ჩვენ ვაგროვებთ ყოველკვირეულ მონაცემებს მთელი 2012 წლისთვის. ყოველ კვირას ვაფიქსირებთ % ცვლილებას აშშ დოლარი/ევროში, % ცვლილებას აშშ-ს ბაზარზე, % ცვლილებას ბრიტანეთის ბაზარზე და % ცვლილებას გერმანიის ბაზარზე.

#### 2) რომელია შემდეგი ალგორითმებისაგან წრფივი და რომელი არაწრფივი და რატომ:
- ლოგისტიკური რეგრესია
- კვადრატული დისკრიმინანტული ანალიზი
- მეოთხე ხარისხის პოლინომიალური რეგრესია:
$$
Y = \beta_0 + \beta_1X_1 + \beta_2X_1X_2 + \beta_3X_1^4 + \beta_4X_1X_2X_3
$$
- წრფივი რეგრესია
- წრფივი დისკრიმინანტული ანალიზი


#### 3) შემდეგი ალგორითმები არის კლასიფიკაციის თუ რეგრესიის:
- ლოგისტიკური რეეგრესია
- მეოთხე ხარისხის პოლინომიალური რეგრესია
- წრფივი დისკრიმინანტული ანალიზი
- კვადრატული დისკრიმინანტული ანალიზი
- Naive Bayes

#### 4) აღწერეთ საშუალო კვადრატული ცდომილების ფორმულა:
$$
RSS = e_1^2 + e_2^2 + \dots e_n^2
$$

#### 5) აჩვენეთ რომ შემდეგი წრფივი რეგრეესსისთვის: 
$$
\hat{y} = \hat{\beta_0} + \hat{\beta_1}X
$$
#### საშუალო კვადრატული გადახრის ცდომილებისათვის:
$$
RSS = e_1^2 + e_2^2 + \dots e_n^2
$$
#### ოპტიმალური პარამეტრები გამოითვლება ფორმულებით:
- გადახრა:
$$
\hat{\beta_1} = \frac{\sum_{i=1}^{n}{(x_i - \bar{x})(y_i - \bar{y})}}{\sum_{i=1}^{n}{(x_i - \bar{x}})^{2}}
$$

- კვეთა:
$$
\hat{\beta_0} = \bar{y} - \bar{b_1}\hat{x}
$$
